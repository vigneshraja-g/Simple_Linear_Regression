```python
#Import the pandas library
import pandas as pd
```

```python
#Load the dataset from a CSV file 
dataset= pd.read_csv("Salary_Data.csv")
```

```python
dataset
```

```python
#Define the independent (input or feature) variable
independent= dataset[["YearsExperience"]]
```

```python
independent
```

```python
## Define the dependent (output or target) variable
dependent= dataset[["Salary"]]
```

```python
dependent
```

```python
#Import the train_test_split function from sklearn to split the dataset
from sklearn.model_selection import train_test_split
## Split the dataset into training and testing sets
X_train, X_test, Y_train, Y_test=train_test_split(independent, dependent, test_size=0.30, random_state=0)
```

```python
X_train
```

```python
X_test
```

```python
#Import the LinearRegression class from sklearn (Procudere)
from sklearn.linear_model import LinearRegression
regressor= LinearRegression()
#Model creation or Learning
regressor.fit(X_train, Y_train)
```

```python
#coefficient (weight)
weight=regressor.coef_
```

```python
weight
```

```python
#intercept (bias)
bais=regressor.intercept_
```

```python
bais
```

```python
#Using the model to tredict test dataset
Y_pred=regressor.predict(X_test)
```

```python
Y_pred
```

```python
#Import the r2_score function from sklearn
from sklearn.metrics import r2_score
#Calculate the R2 score
r_score= r2_score(Y_test, Y_pred)
```

```python
r_score
```

```python
#Import the pickle library to save and load the trained model
import pickle
filename="finalized_model_linear.sav"
```

```python
#Save the trained model to a file using pickle
pickle.dump(regressor, open(filename,"wb"))
```

```python
# Load the saved model from the file
loaded_model=pickle.load(open("finalized_model_linear.sav","rb"))
result=loaded_model.predict([[15]])
```
